## DIFFERENT REGRESSIONS ON BOSTON DATA SET

```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = load_boston()
```

```python
df
```

```python
# Get the data and store it as a DataFrame
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
#Get columns from "feature_names" in dataset and set columns
dataset.columns = df.feature_names
```

```python
dataset
```

```python
#Setting target value
dataset["Price"] = df.target
```

```python
dataset
```

```python
#Identifying dependent and independent variables:
X = dataset.iloc[:,:-1] #Till -1 index column (Independent)
y = dataset.iloc[:,-1] #Only -1 index i.e Sales prices (Dependent)
```

## Linear Regression
We will perform Linear regression on the above data

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lin_regressor = LinearRegression()
cvscore = cross_val_score(lin_regressor,X,y,scoring = "neg_mean_squared_error",cv = 10)
#Basically means we are doing cross validation on the linear model object with X and y as the data
#Scoring method in negative mean squared error(The closer to zero the better)
#cv it the number of experiments we will do for cross validation(The more number, the closer output to zero)
print("CV Scores for experiments:{}".format(cvscore)) #Show score for each cv experiment (10 here since value of cv=10)
mean_cvscore = np.mean(cvscore) #Doing mean of all values of cv
print("Mean CV Score:{}".format(mean_cvscore))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
parameters = {"alpha":[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,15,20,30,35,40,45,50,55,60,100,150,200,300]} 
ridge_regressor = GridSearchCV(ridge,parameters,scoring="neg_mean_squared_error",cv=10)
#GridSearchCV will help us find out which Alpha value will be suitable for our ridge regressor
#Once identified, we can fit our regressor model
ridge_regressor.fit(X,y)
```

```python
#Lets print the Parameter for Alpha chosen
print("Parameter chosen:{}".format(ridge_regressor.best_params_))
print("Mean Score:{}".format(ridge_regressor.best_score_))
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
parameters = {"alpha":[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,15,20,30,35,40,45,50,55,60,100,150,200,300]}
lasso_regressor = GridSearchCV(lasso,parameters,scoring = "neg_mean_squared_error",cv=10)
lasso_regressor.fit(X,y)
```

```python
print("Best Parameter:{}".format(lasso_regressor.best_params_))
print("Best Score:{}".format(lasso_regressor.best_score_))
```

# Comparing all Negative Mean Squared Score for Train Data:

```python
print("Linear Regressor:{}".format(mean_cvscore))
print("Ridge Regressor:{}".format(ridge_regressor.best_score_))
print("Lasso Regressor:{}".format(lasso_regressor.best_score_))
```
